```sql
USE CATALOG smart_odoo;

-- =====================================================
-- Table: silver.crm_stage (2)
-- =====================================================
 
CREATE TABLE IF NOT EXISTS silver.crm_stage
(
    crm_stage_id INT,
 
    sequence INT,
    rotting_threshold_days INT,
    name STRING,
    requirements STRING,
 
    is_won BOOLEAN,
    fold BOOLEAN,
 
    created_at TIMESTAMP,
    updated_at TIMESTAMP,
 
    dwh_loaded_at TIMESTAMP,
    dwh_source_db STRING
)
USING DELTA;
 
-- MERGE
 
MERGE INTO silver.crm_stage AS target
USING
(
    SELECT
        CAST(cs.id AS INT)                      AS crm_stage_id,
        CAST(cs.sequence AS INT)                AS sequence,
        CAST(cs.rotting_threshold_days AS INT)  AS rotting_threshold_days,
        cs.name,
        cs.requirements,
        CAST(cs.is_won AS BOOLEAN) AS is_won,
        CAST(cs.fold AS BOOLEAN)   AS fold,
        CAST(cs.create_date AS TIMESTAMP) AS created_at,
        CAST(cs.write_date AS TIMESTAMP)  AS updated_at,
        current_timestamp() AS dwh_loaded_at,
        cs.dwh_source_db
 
    FROM bronze.crm_stage cs
    WHERE CAST(cs.write_date AS TIMESTAMP) > COALESCE
    (
        (SELECT last_write_date FROM silver.load_tracking WHERE table_name = 'crm_stage'),
        CAST('1900-01-01' AS TIMESTAMP)
    )
) AS source
 
ON target.crm_stage_id = source.crm_stage_id
 
WHEN MATCHED THEN UPDATE SET
    target.sequence = source.sequence,
    target.rotting_threshold_days = source.rotting_threshold_days,
    target.name = source.name,
    target.requirements = source.requirements,
    target.is_won = source.is_won,
    target.fold = source.fold,
    target.created_at = source.created_at,
    target.updated_at = source.updated_at,
    target.dwh_loaded_at = current_timestamp(),
    target.dwh_source_db = source.dwh_source_db
 
WHEN NOT MATCHED THEN INSERT *;
 
```
