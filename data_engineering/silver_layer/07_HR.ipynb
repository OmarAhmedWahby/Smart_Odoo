```sql
USE CATALOG smart_odoo;
 
-- =====================================================
-- Table: silver.hr_employee (1)
-- =====================================================
 
CREATE TABLE IF NOT EXISTS silver.hr_employee
(
    hr_employee_id INT,
 
    company_id INT,
    company_name STRING,
 
    user_id INT,
    user_name STRING,
 
    parent_id INT,
    parent_name STRING,
 
    coach_id INT,
    coach_name STRING,
 
    work_contact_id INT,
    work_contact_name STRING,
 
    name STRING,
    legal_name STRING,
    work_phone STRING,
    mobile_phone STRING,
    work_email STRING,
    lang STRING,
 
    certificate STRING,
    study_field STRING,
    study_school STRING,
    place_of_birth STRING,
 
    emergency_contact STRING,
    emergency_phone STRING,
 
    permit_no STRING,
    visa_no STRING,
    barcode STRING,
    today_location_name STRING,
 
    salary_distribution STRING,
    employee_properties STRING,
 
    active BOOLEAN,
    birthday_public_display BOOLEAN,
    work_permit_scheduled_activity BOOLEAN,
 
    birthday TIMESTAMP,
    visa_expire TIMESTAMP,
    work_permit_expiration_date TIMESTAMP,
 
    created_at TIMESTAMP,
    updated_at TIMESTAMP,
 
    dwh_loaded_at TIMESTAMP,
    dwh_source_db STRING
)
USING DELTA;
 
-- MERGE
 
MERGE INTO silver.hr_employee AS target
USING
(
    SELECT
        CAST(e.id AS INT) AS hr_employee_id,
 
        CAST(GET_JSON_OBJECT(e.company_id, '$[0]') AS INT) AS company_id,
        GET_JSON_OBJECT(e.company_id, '$[1]')               AS company_name,
 
        CAST(GET_JSON_OBJECT(e.user_id, '$[0]') AS INT)    AS user_id,
        GET_JSON_OBJECT(e.user_id, '$[1]')                  AS user_name,
 
        CAST(GET_JSON_OBJECT(e.parent_id, '$[0]') AS INT)  AS parent_id,
        GET_JSON_OBJECT(e.parent_id, '$[1]')                AS parent_name,
 
        CAST(GET_JSON_OBJECT(e.coach_id, '$[0]') AS INT)   AS coach_id,
        GET_JSON_OBJECT(e.coach_id, '$[1]')                 AS coach_name,
 
        CAST(GET_JSON_OBJECT(e.work_contact_id, '$[0]') AS INT) AS work_contact_id,
        GET_JSON_OBJECT(e.work_contact_id, '$[1]')               AS work_contact_name,
 
        e.name,
        e.legal_name,
        e.work_phone,
        e.mobile_phone,
        e.work_email,
        e.lang,
 
        e.certificate,
        e.study_field,
        e.study_school,
        e.place_of_birth,
 
        e.emergency_contact,
        e.emergency_phone,
 
        e.permit_no,
        e.visa_no,
        e.barcode,
        e.today_location_name,
 
        e.salary_distribution,
        e.employee_properties,
 
        CAST(e.active AS BOOLEAN)                          AS active,
        CAST(e.birthday_public_display AS BOOLEAN)         AS birthday_public_display,
        CAST(e.work_permit_scheduled_activity AS BOOLEAN)  AS work_permit_scheduled_activity,
 
        CAST(e.birthday AS TIMESTAMP)                      AS birthday,
        CAST(e.visa_expire AS TIMESTAMP)                   AS visa_expire,
        CAST(e.work_permit_expiration_date AS TIMESTAMP)   AS work_permit_expiration_date,
 
        CAST(e.create_date AS TIMESTAMP) AS created_at,
        CAST(e.write_date AS TIMESTAMP)  AS updated_at,
 
        current_timestamp() AS dwh_loaded_at,
        e.dwh_source_db
 
    FROM bronze.hr_employee e
    WHERE CAST(e.write_date AS TIMESTAMP) > COALESCE
    (
        (SELECT last_write_date FROM silver.load_tracking WHERE table_name = 'hr_employee'),
        CAST('1900-01-01' AS TIMESTAMP)
    )
) AS source
 
ON target.hr_employee_id = source.hr_employee_id
 
WHEN MATCHED THEN UPDATE SET
    target.company_id                      = source.company_id,
    target.company_name                    = source.company_name,
    target.user_id                         = source.user_id,
    target.user_name                       = source.user_name,
    target.parent_id                       = source.parent_id,
    target.parent_name                     = source.parent_name,
    target.coach_id                        = source.coach_id,
    target.coach_name                      = source.coach_name,
    target.work_contact_id                 = source.work_contact_id,
    target.work_contact_name               = source.work_contact_name,
    target.name                            = source.name,
    target.legal_name                      = source.legal_name,
    target.work_phone                      = source.work_phone,
    target.mobile_phone                    = source.mobile_phone,
    target.work_email                      = source.work_email,
    target.lang                            = source.lang,
    target.certificate                     = source.certificate,
    target.study_field                     = source.study_field,
    target.study_school                    = source.study_school,
    target.place_of_birth                  = source.place_of_birth,
    target.emergency_contact               = source.emergency_contact,
    target.emergency_phone                 = source.emergency_phone,
    target.permit_no                       = source.permit_no,
    target.visa_no                         = source.visa_no,
    target.barcode                         = source.barcode,
    target.today_location_name             = source.today_location_name,
    target.salary_distribution             = source.salary_distribution,
    target.employee_properties             = source.employee_properties,
    target.active                          = source.active,
    target.birthday_public_display         = source.birthday_public_display,
    target.work_permit_scheduled_activity  = source.work_permit_scheduled_activity,
    target.birthday                        = source.birthday,
    target.visa_expire                     = source.visa_expire,
    target.work_permit_expiration_date     = source.work_permit_expiration_date,
    target.created_at                      = source.created_at,
    target.updated_at                      = source.updated_at,
    target.dwh_loaded_at                   = current_timestamp(),
    target.dwh_source_db                   = source.dwh_source_db
 
WHEN NOT MATCHED THEN INSERT *;
 
```

```sql
USE CATALOG smart_odoo;
 
-- =====================================================
-- Table: silver.hr_department (2)
-- =====================================================
 
CREATE TABLE IF NOT EXISTS silver.hr_department
(
    hr_department_id INT,
 
    company_id INT,
    company_name STRING,
 
    parent_id INT,
    parent_name STRING,
 
    manager_id INT,
    manager_name STRING,
 
    master_department_id INT,
    master_department_name STRING,
 
    name STRING,
    parent_path STRING,
    note STRING,
 
    active BOOLEAN,
 
    created_at TIMESTAMP,
    updated_at TIMESTAMP,
 
    dwh_loaded_at TIMESTAMP,
    dwh_source_db STRING
)
USING DELTA;
 
-- MERGE
 
MERGE INTO silver.hr_department AS target
USING
(
    SELECT
        CAST(d.id AS INT) AS hr_department_id,
 
        CAST(GET_JSON_OBJECT(d.company_id, '$[0]') AS INT)            AS company_id,
        GET_JSON_OBJECT(d.company_id, '$[1]')                          AS company_name,
 
        CAST(GET_JSON_OBJECT(d.parent_id, '$[0]') AS INT)             AS parent_id,
        GET_JSON_OBJECT(d.parent_id, '$[1]')                           AS parent_name,
 
        CAST(GET_JSON_OBJECT(d.manager_id, '$[0]') AS INT)            AS manager_id,
        GET_JSON_OBJECT(d.manager_id, '$[1]')                          AS manager_name,
 
        CAST(GET_JSON_OBJECT(d.master_department_id, '$[0]') AS INT)  AS master_department_id,
        GET_JSON_OBJECT(d.master_department_id, '$[1]')                AS master_department_name,
 
        d.name,
        d.parent_path,
        d.note,
 
        CAST(d.active AS BOOLEAN) AS active,
 
        CAST(d.create_date AS TIMESTAMP) AS created_at,
        CAST(d.write_date AS TIMESTAMP)  AS updated_at,
 
        current_timestamp() AS dwh_loaded_at,
        d.dwh_source_db
 
    FROM bronze.hr_department d
    WHERE CAST(d.write_date AS TIMESTAMP) > COALESCE
    (
        (SELECT last_write_date FROM silver.load_tracking WHERE table_name = 'hr_department'),
        CAST('1900-01-01' AS TIMESTAMP)
    )
) AS source
 
ON target.hr_department_id = source.hr_department_id
 
WHEN MATCHED THEN UPDATE SET
    target.company_id              = source.company_id,
    target.company_name            = source.company_name,
    target.parent_id               = source.parent_id,
    target.parent_name             = source.parent_name,
    target.manager_id              = source.manager_id,
    target.manager_name            = source.manager_name,
    target.master_department_id    = source.master_department_id,
    target.master_department_name  = source.master_department_name,
    target.name                    = source.name,
    target.parent_path             = source.parent_path,
    target.note                    = source.note,
    target.active                  = source.active,
    target.created_at              = source.created_at,
    target.updated_at              = source.updated_at,
    target.dwh_loaded_at           = current_timestamp(),
    target.dwh_source_db           = source.dwh_source_db
 
WHEN NOT MATCHED THEN INSERT *;
```

```sql
USE CATALOG smart_odoo;
 
-- =====================================================
-- Table: silver.hr_job (3)
-- =====================================================
 
CREATE TABLE IF NOT EXISTS silver.hr_job
(
    hr_job_id INT,
 
    company_id INT,
    company_name STRING,
 
    department_id INT,
    department_name STRING,
 
    user_id INT,
    user_name STRING,
 
    contract_type_id INT,
    contract_type_name STRING,
 
    sequence INT,
    no_of_recruitment INT,
 
    name STRING,
    description STRING,
    requirements STRING,
 
    active BOOLEAN,
 
    created_at TIMESTAMP,
    updated_at TIMESTAMP,
 
    dwh_loaded_at TIMESTAMP,
    dwh_source_db STRING
)
USING DELTA;
 
-- MERGE
 
MERGE INTO silver.hr_job AS target
USING
(
    SELECT
        CAST(j.id AS INT) AS hr_job_id,
 
        CAST(GET_JSON_OBJECT(j.company_id, '$[0]') AS INT)       AS company_id,
        GET_JSON_OBJECT(j.company_id, '$[1]')                     AS company_name,
 
        CAST(GET_JSON_OBJECT(j.department_id, '$[0]') AS INT)    AS department_id,
        GET_JSON_OBJECT(j.department_id, '$[1]')                  AS department_name,
 
        CAST(GET_JSON_OBJECT(j.user_id, '$[0]') AS INT)          AS user_id,
        GET_JSON_OBJECT(j.user_id, '$[1]')                        AS user_name,
 
        CAST(GET_JSON_OBJECT(j.contract_type_id, '$[0]') AS INT) AS contract_type_id,
        GET_JSON_OBJECT(j.contract_type_id, '$[1]')               AS contract_type_name,
 
        CAST(j.sequence AS INT)          AS sequence,
        CAST(j.no_of_recruitment AS INT) AS no_of_recruitment,
 
        j.name,
        j.description,
        j.requirements,
 
        CAST(j.active AS BOOLEAN) AS active,
 
        CAST(j.create_date AS TIMESTAMP) AS created_at,
        CAST(j.write_date AS TIMESTAMP)  AS updated_at,
 
        current_timestamp() AS dwh_loaded_at,
        j.dwh_source_db
 
    FROM bronze.hr_job j
    WHERE CAST(j.write_date AS TIMESTAMP) > COALESCE
    (
        (SELECT last_write_date FROM silver.load_tracking WHERE table_name = 'hr_job'),
        CAST('1900-01-01' AS TIMESTAMP)
    )
) AS source
 
ON target.hr_job_id = source.hr_job_id
 
WHEN MATCHED THEN UPDATE SET
    target.company_id          = source.company_id,
    target.company_name        = source.company_name,
    target.department_id       = source.department_id,
    target.department_name     = source.department_name,
    target.user_id             = source.user_id,
    target.user_name           = source.user_name,
    target.contract_type_id    = source.contract_type_id,
    target.contract_type_name  = source.contract_type_name,
    target.sequence            = source.sequence,
    target.no_of_recruitment   = source.no_of_recruitment,
    target.name                = source.name,
    target.description         = source.description,
    target.requirements        = source.requirements,
    target.active              = source.active,
    target.created_at          = source.created_at,
    target.updated_at          = source.updated_at,
    target.dwh_loaded_at       = current_timestamp(),
    target.dwh_source_db       = source.dwh_source_db
 
WHEN NOT MATCHED THEN INSERT *;
```
