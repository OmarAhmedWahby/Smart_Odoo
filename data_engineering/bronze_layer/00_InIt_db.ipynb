```sql
CREATE  CATALOG if NOT EXISTS smart_odoo;
USE CATALOG smart_odoo;

CREATE SCHEMA IF NOT EXISTS smart_odoo.bronze;
CREATE SCHEMA IF NOT EXISTS smart_odoo.silver;
CREATE SCHEMA IF NOT EXISTS smart_odoo.gold;


Drop TABLE IF EXISTS bronze.load_tracking;
DROP TABLE IF EXISTS silver.load_tracking;

CREATE TABLE IF NOT EXISTS bronze.load_tracking (
    table_name STRING,
    last_write_date TIMESTAMP,
    PRIMARY KEY (table_name)
);

CREATE TABLE IF NOT EXISTS silver.load_tracking (
    table_name STRING,
    last_write_date TIMESTAMP,
    PRIMARY KEY (table_name)
);
```
