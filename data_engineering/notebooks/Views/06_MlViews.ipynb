```sql
use catalog smart_odoo;
use schema gold;

CREATE OR REPLACE VIEW vm_ml_inventory_stockout_prediction AS

Select product_id , product_name , available_qty , days_of_stock , stockout_probability , will_stockout , risk_level  from smart_odoo.gold.ml_inventory_stockout_predictions
ORDER BY  stockout_probability DESC 
limit 5 ;
```

```sql
use catalog smart_odoo;
use schema gold;

CREATE OR REPLACE VIEW vm_ml_invoice_delay_scoring AS
Select partner_id , delay_probability , risk_bucket from smart_odoo.gold.ml_invoice_delay_scoring
ORDER BY delay_probability DESC

limit 10 ;
```

```sql
USE CATALOG smart_odoo;
USE SCHEMA gold;

CREATE OR REPLACE VIEW vm_ml_leads_scores AS

SELECT
    lead_id,
    lead_name,
    probability_of_win,
    expected_revenue,
    lead_score
FROM smart_odoo.gold.ml_lead_scores
WHERE expected_revenue <> 0  
ORDER BY probability_of_win DESC, expected_revenue DESC
LIMIT 5;

```

```sql
USE CATALOG smart_odoo;
USE SCHEMA gold;

CREATE OR REPLACE VIEW vm_ml_overdue_GrowthCustomer AS

SELECT
    partner_name,
    segment,
    paid_ratio,
    overdue_ratio,
    collection_rate
FROM smart_odoo.gold.ml_customer_segments
 WHERE segment = "🔄 Developing — Growing Relationship"
ORDER BY overdue_ratio DESC, paid_ratio DESC
LIMIT 5;
```

```sql
USE CATALOG smart_odoo;
USE SCHEMA gold;

CREATE OR REPLACE VIEW vm_ml_overdue_RiskCustomer AS

SELECT
    partner_name,
    segment,
    paid_ratio,
    overdue_ratio,
    collection_rate
FROM smart_odoo.gold.ml_customer_segments
 WHERE segment = "⚠️ At-Risk — Frequent Delays"
ORDER BY overdue_ratio DESC, paid_ratio DESC
LIMIT 5;
```

```sql
USE CATALOG smart_odoo;
USE SCHEMA gold;

CREATE OR REPLACE VIEW vm_ml_overdue_DormantCustomer AS

SELECT
    partner_name,
    segment,
    paid_ratio,
    overdue_ratio,
    collection_rate
FROM smart_odoo.gold.ml_customer_segments
 WHERE segment = "💤 Dormant — Low Engagement"
ORDER BY overdue_ratio DESC, paid_ratio DESC
LIMIT 5;
```

```sql
USE CATALOG smart_odoo;
USE SCHEMA gold;

CREATE OR REPLACE VIEW vm_ml_cancelled_Customer AS

SELECT
    partner_name,
    segment,
    paid_ratio,
    cancel_ratio,
    collection_rate
FROM smart_odoo.gold.ml_customer_segments
ORDER BY cancel_ratio DESC, paid_ratio DESC
LIMIT 5;
```
