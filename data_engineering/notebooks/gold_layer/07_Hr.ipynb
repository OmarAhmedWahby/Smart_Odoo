```sql
USE CATALOG smart_odoo;
 
-- =====================================================
-- DIM: gold.dim_employee
-- =====================================================
 
CREATE TABLE IF NOT EXISTS gold.dim_employee
(
    employee_id INT,
 
    employee_name STRING,
    legal_name STRING,
 
    company_id INT,
    company_name STRING,
 
    manager_id INT,
    manager_name STRING,
 
    coach_id INT,
    coach_name STRING,
 
    work_email STRING,
    work_phone STRING,
    mobile_phone STRING,
 
    certificate STRING,
    study_field STRING,
    study_school STRING,
 
    birthday TIMESTAMP,
    visa_expire TIMESTAMP,
    work_permit_expiration_date TIMESTAMP,
 
    is_active BOOLEAN,
 
    created_at TIMESTAMP,
    updated_at TIMESTAMP,
 
    dwh_loaded_at TIMESTAMP
)
USING DELTA;
 
-- MERGE
 
MERGE INTO gold.dim_employee AS target
 
USING
(
    SELECT
        hr_employee_id      AS employee_id,
        name                AS employee_name,
        legal_name,
        company_id,
        company_name,
        parent_id           AS manager_id,
        parent_name         AS manager_name,
        coach_id,
        coach_name,
        work_email,
        work_phone,
        mobile_phone,
        certificate,
        study_field,
        study_school,
        birthday,
        visa_expire,
        work_permit_expiration_date,
        active              AS is_active,
        created_at,
        updated_at,
        current_timestamp() AS dwh_loaded_at
    FROM silver.hr_employee
    WHERE updated_at > COALESCE
    (
        (SELECT MAX(updated_at) FROM gold.dim_employee),
        CAST('1900-01-01' AS TIMESTAMP)
    )
) AS source
 
ON target.employee_id = source.employee_id
 
WHEN MATCHED AND target.updated_at < source.updated_at THEN
UPDATE SET
    target.employee_name                = source.employee_name,
    target.legal_name                   = source.legal_name,
    target.company_id                   = source.company_id,
    target.company_name                 = source.company_name,
    target.manager_id                   = source.manager_id,
    target.manager_name                 = source.manager_name,
    target.coach_id                     = source.coach_id,
    target.coach_name                   = source.coach_name,
    target.work_email                   = source.work_email,
    target.work_phone                   = source.work_phone,
    target.mobile_phone                 = source.mobile_phone,
    target.certificate                  = source.certificate,
    target.study_field                  = source.study_field,
    target.study_school                 = source.study_school,
    target.birthday                     = source.birthday,
    target.visa_expire                  = source.visa_expire,
    target.work_permit_expiration_date  = source.work_permit_expiration_date,
    target.is_active                    = source.is_active,
    target.updated_at                   = source.updated_at,
    target.dwh_loaded_at                = current_timestamp()
 
WHEN NOT MATCHED THEN INSERT *;
```

```sql
USE CATALOG smart_odoo;
 
-- =====================================================
-- FACT: gold.fact_employee_snapshot
-- =====================================================
 
CREATE TABLE IF NOT EXISTS gold.fact_employee_snapshot
(
    snapshot_date DATE,
 
    employee_id INT,
    employee_name STRING,
 
    department_id INT,
    department_name STRING,
 
    job_id INT,
    job_name STRING,
 
    company_id INT,
    company_name STRING,
 
    manager_id INT,
    manager_name STRING,
 
    is_active BOOLEAN,
    headcount INT,
 
    dwh_loaded_at TIMESTAMP
)
USING DELTA;
 
-- MERGE
 
MERGE INTO gold.fact_employee_snapshot AS target
 
USING
(
    SELECT
        CAST(current_date() AS DATE)    AS snapshot_date,
 
        e.hr_employee_id                AS employee_id,
        e.name                          AS employee_name,
 
        d.hr_department_id              AS department_id,
        d.name                          AS department_name,
 
        j.hr_job_id                     AS job_id,
        j.name                          AS job_name,
 
        e.company_id,
        e.company_name,
 
        e.parent_id                     AS manager_id,
        e.parent_name                   AS manager_name,
 
        e.active                        AS is_active,
        1                               AS headcount,
 
        current_timestamp()             AS dwh_loaded_at
 
    FROM silver.hr_employee e
    LEFT JOIN silver.hr_department d
        ON d.hr_department_id = e.hr_employee_id  
    LEFT JOIN silver.hr_job j
        ON j.department_id = d.hr_department_id
        AND j.company_id   = e.company_id
 
) AS source
 
ON  target.snapshot_date = source.snapshot_date
AND target.employee_id   = source.employee_id
 
WHEN MATCHED THEN
UPDATE SET
    target.department_id    = source.department_id,
    target.department_name  = source.department_name,
    target.job_id           = source.job_id,
    target.job_name         = source.job_name,
    target.company_id       = source.company_id,
    target.company_name     = source.company_name,
    target.manager_id       = source.manager_id,
    target.manager_name     = source.manager_name,
    target.is_active        = source.is_active,
    target.headcount        = source.headcount,
    target.dwh_loaded_at    = current_timestamp()
 
WHEN NOT MATCHED THEN INSERT *;
```
