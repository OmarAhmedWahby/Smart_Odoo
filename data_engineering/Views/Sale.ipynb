```sql
use catalog smart_odoo;
use schema gold;

CREATE OR REPLACE view gold.sales_kpis AS
SELECT
    ROUND(SUM(net_revenue), 2) AS total_sales
FROM gold.fact_sales;

```

```sql
Select total_sales from smart_odoo.gold.sales_kpis;
```

```sql
Select sum(net_revenue)
from smart_odoo.gold.fact_sales
where order_state = 'no'
```
